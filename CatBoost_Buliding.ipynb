# In OOP Shape

```python
class Data_Prep:
    def __init__(self,Link=link) -> None:
        self.Link=Link
        self.Data=read_data(self.Link)
        self.X = Data.drop(columns=['Price'])  # All columns except 'Price' are features
        self.y= Data['Price']  # Target variable is 'Price'

        pass


    def read_data(self):
        Data=pd.read_csv(self.Link,index_col=0)
        Data.drop_duplicates(inplace=True)
        return Data
    
    def Normalize_Data(self):
        scaler = MinMaxScaler()
        normalized_data = scaler.fit_transform(self.Data)
        return normalized_data
    
    def Spliter (self):
        X_=self.Normalize_Data()

        X_train, X_test, y_train, y_test = train_test_split(X_, self.y, test_size=0.2, random_state=42)
        return X_train, X_test, y_train, y_test
```

```python
Data_obj=Data_Prep()
```

```python
X_train, X_test, y_train, y_test =Data_obj.Spliter()
```

```python
len(y_train)
```

# The Model

```python
# Initialize CatBoostRegressor
model = CatBoostRegressor(iterations=1000,  # Number of boosting iterations
                          learning_rate=0.1,  # Learning rate
                          depth=6,  # Depth of the trees
                          random_seed=42  # Random seed for reproducibility  
                          )
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = model.predict(X_test)
#
# Calculate RÂ² score
r2 = r2_score(y_test, y_pred)
print(f'RÂ² score: {r2}')
```
